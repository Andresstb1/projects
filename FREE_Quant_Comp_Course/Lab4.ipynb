## **🔹 2. Implementación de la teleportación cuántica**  

La **teleportación cuántica** es uno de los conceptos más fascinantes de la computación cuántica. Aunque el nombre sugiere una conexión con la ciencia ficción, en realidad es un fenómeno físico real basado en **el entrelazamiento cuántico** y la **transferencia de información cuántica sin transmisión física de partículas**.  

### **📌 ¿Cómo funciona la teleportación cuántica?**  
El protocolo de **teleportación cuántica** permite transferir el estado de un qubit **A** a un qubit **C**, aunque estos qubits nunca hayan interactuado directamente.  

Se basa en tres pasos clave:  
1️⃣ **Creación de un par de qubits entrelazados** entre **B** y **C**.  
2️⃣ **Medición conjunta de los qubits A y B** y el envío de **bits clásicos** con el resultado de la medición.  
3️⃣ **Corrección del estado del qubit C** usando operaciones clásicas condicionadas para obtener el estado original de A en C.  

💡 **Aplicaciones del protocolo de teleportación:**  
🔹 **Seguridad cuántica y criptografía**: La teleportación cuántica es clave en la transmisión segura de información.  
🔹 **Computación cuántica distribuida**: Permite conectar computadoras cuánticas a través de **redes cuánticas**.  
🔹 **Corrección de errores cuánticos**: Ayuda a mantener la integridad de la información cuántica.

# **📌 Implementación de la Teleportación Cuántica**

En este apartado, exploraremos en profundidad **el protocolo de teleportación cuántica**, uno de los experimentos más fascinantes en computación cuántica. Analizaremos su **teoría**, su **implementación en Qiskit**, y estudiaremos su **ejecución y resultados**.

---

## **1️⃣ Teoría de la Teleportación Cuántica**  

### **🔹 ¿Qué es la teleportación cuántica?**  
La **teleportación cuántica** permite transferir el **estado de un qubit** a otro **sin que exista una conexión física directa entre ellos**. Esta técnica se basa en **el entrelazamiento cuántico** y el uso de **bits clásicos para la transmisión de información**.

💡 **Importante:**  
- No se trata de mover físicamente un qubit de un lugar a otro.  
- La información cuántica se transfiere a través de **qubits entrelazados y comunicación clásica**.  
- Es clave para **redes cuánticas, corrección de errores y comunicaciones cuánticas seguras**.

---

### **🔹 Concepto Matemático**
Supongamos que queremos **transferir el estado de un qubit $A$** a otro qubit **$C$** usando un **qubit auxiliar $B$**.  

1️⃣ **Estado inicial del qubit $A$** (desconocido):  
$$
|\psi\rangle_A = \alpha |0\rangle + \beta |1\rangle
$$
2️⃣ **Entrelazamiento entre los qubits $B$ y $C$**:  
$$
|\Phi^+\rangle_{BC} = \frac{1}{\sqrt{2}} (|00\rangle_{BC} + |11\rangle_{BC})
$$
3️⃣ **El sistema total es la combinación de ambos**:  
$$
|\psi\rangle_A \otimes |\Phi^+\rangle_{BC}
$$
4️⃣ **Medición conjunta en $A$ y $B$** y transmisión de bits clásicos.  
5️⃣ **Corrección en $C$ mediante operaciones condicionales**, recuperando $|\psi\rangle$ en **$C$**.

---

### **🔹 Pasos del Protocolo de Teleportación**  

1️⃣ **Generar un par de qubits entrelazados** entre **B** y **C**.  
2️⃣ **Aplicar una medición Bell a los qubits A y B**.  
3️⃣ **Enviar el resultado de la medición a través de un canal clásico**.  
4️⃣ **Aplicar operaciones cuánticas en C basadas en la medición** para reconstruir el estado original.

---

## **2️⃣ Implementación en Qiskit**

### **📌 Explicación del Circuito**
📌 Para implementar la teleportación cuántica en **Qiskit**, seguimos estos pasos:

1️⃣ **Creamos 3 qubits**:  
   - **Qubit A:** Contiene el estado a teletransportar.  
   - **Qubit B y Qubit C:** Están entrelazados para permitir la teleportación.  

2️⃣ **Preparamos el entrelazamiento entre los qubits B y C** usando una **puerta Hadamard (H) y CNOT**.  

3️⃣ **Realizamos una medición Bell** en los qubits A y B.  

4️⃣ **Usamos los bits clásicos para corregir el estado en C** con **puertas X y Z condicionales**.

---

### **📌 Código en Qiskit**

```python
# Importar librerías necesarias
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram, array_to_latex
from qiskit.quantum_info import Statevector

# Simulador cuántico
simulator = Aer.get_backend('aer_simulator')

# Crear registros cuánticos y clásicos
qr = QuantumRegister(3, 'q')  # 3 qubits
cr = ClassicalRegister(2, 'c')  # 2 bits clásicos
qc = QuantumCircuit(qr, cr)

# -------------------------------
# 🔹 Paso 1: Preparar el estado a teletransportar
# -------------------------------
qc.initialize([0.6, 0.8], 0)  # Estado arbitrario en qubit A

# -------------------------------
# 🔹 Paso 2: Crear el entrelazamiento entre B y C
# -------------------------------
qc.h(1)  # Hadamard en el qubit B
qc.cx(1, 2)  # CNOT entre qubit B y C

# -------------------------------
# 🔹 Paso 3: Medición Bell en qubits A y B
# -------------------------------
qc.cx(0, 1)  # CNOT de A a B
qc.h(0)  # Hadamard en A

qc.measure(0, 0)  # Medir A
qc.measure(1, 1)  # Medir B

# -------------------------------
# 🔹 Paso 4: Corrección del estado en qubit C
# -------------------------------
qc.x(2).c_if(cr, 1)  # Aplicar X si el bit clásico 1 es 1
qc.z(2).c_if(cr, 2)  # Aplicar Z si el bit clásico 0 es 1

# Dibujar el circuito
qc.draw('mpl')
```

## **3️⃣ Ejecución y Análisis de Resultados**

### **📌 Ejecución en el Simulador**

```python
# Transpilar el circuito y ejecutarlo en el simulador
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)

# Obtener resultados de medición
result = job.result()
counts = result.get_counts(qc)

# Mostrar histograma de resultados
plot_histogram(counts)
```

### **📌 Estado Final de Qubit C (Verificación)**

```python
# -------------------------------
# 🔹 Análisis del Estado Antes de la Medición
# -------------------------------
# Creamos una versión sin mediciones para calcular el estado cuántico
qc_sin_medicion = qc.remove_final_measurements(inplace=False)

# Obtener el estado cuántico antes de la medición
estado_final = Statevector(qc_sin_medicion)
print("\nEstado final antes de la medición:")
display(array_to_latex(estado_final.data, prefix="|ψ⟩ = "))

# -------------------------------
# 🔹 Ejecución en el Simulador
# -------------------------------
# Transpilar el circuito y ejecutarlo en el simulador
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)

# Obtener resultados de medición
result = job.result()
counts = result.get_counts(qc)

# Mostrar histograma de resultados
plot_histogram(counts)
```

# **🔹 Algoritmo de Deutsch-Jozsa**
El **Algoritmo de Deutsch-Jozsa** fue uno de los primeros en demostrar una **ventaja cuántica clara** sobre los algoritmos clásicos.

### **📌 ¿Cuál es el Problema?**
Tenemos una **función $f(x)$** que acepta un **bit de entrada (x)** y devuelve **0 o 1**.  

Sabemos que la función es:  
1️⃣ **Constante:** Siempre devuelve el mismo valor para cualquier entrada.  
2️⃣ **Balanceada:** Devuelve 0 la mitad de las veces y 1 la otra mitad.  

**Objetivo:**  
📌 **Determinar si la función es constante o balanceada con solo UNA evaluación**.  

### **📌 Implementación en Qiskit**


```python
# Importar librerías necesarias con Qiskit 1.0+
from qiskit.circuit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram

# Crear el circuito cuántico para Deutsch-Jozsa con 2 qubits
qc = QuantumCircuit(2, 1)

# Inicializar en superposición usando Hadamard
qc.h([0, 1])

# Aplicar una puerta CNOT con el primer qubit como control y el segundo como objetivo
qc.cx(0, 1)

# Aplicar Hadamard nuevamente
qc.h(0)

# Medir el primer qubit
qc.measure(0, 0)

# Dibujar el circuito
qc.draw('mpl')

# Simulador cuántico
simulator = Aer.get_backend('aer_simulator')
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)
result = job.result()

# Obtener los resultados
counts = result.get_counts(qc)
plot_histogram(counts)
```

📌 **Resultados esperados:**  
✅ **Si la función es constante**, siempre obtenemos **"0"**.  
✅ **Si la función es balanceada**, obtenemos **"1"**.  

✅ **Supera a cualquier algoritmo clásico, que necesitaría múltiples evaluaciones para verificar todas las entradas.**

# **🔹 Algoritmo de Grover**
El **Algoritmo de Grover** resuelve el problema de **búsqueda en bases de datos desordenadas** con **una mejora cuadrática** sobre los algoritmos clásicos.

### **📌 ¿Cuál es el Problema?**
Dada una **base de datos de $N$ elementos**, queremos encontrar un **elemento marcado**.  

📌 **En una computadora clásica, esto tomaría $\mathcal{O}(N)$ intentos** en promedio.  
📌 **Grover lo reduce a $\mathcal{O}(\sqrt{N})$ intentos** usando **amplificación de amplitud**.  

---

### **📌 Implementación en Qiskit**

```python
# Importar librerías necesarias con Qiskit 1.0+
from qiskit.circuit import QuantumCircuit
from qiskit_aer import Aer  # ✅ Importación corregida para Qiskit 1.0+
from qiskit import transpile
from qiskit.visualization import plot_histogram

# Crear un circuito con 2 qubits y 2 bits clásicos
qc = QuantumCircuit(2, 2)

# Inicializar en superposición
qc.h([0, 1])

# Aplicar una puerta X en el segundo qubit (marcar un estado)
qc.x(1)

# Aplicar una compuerta Z controlada para amplificar el estado marcado
qc.cz(0, 1)

# Aplicar Hadamard nuevamente
qc.h([0, 1])

# Medir ambos qubits
qc.measure([0, 1], [0, 1])

# Dibujar el circuito
qc.draw('mpl')

# Ejecutar el circuito en el simulador cuántico
simulator = Aer.get_backend('aer_simulator')
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)
result = job.result()

# Obtener y visualizar los resultados
counts = result.get_counts(qc)
plot_histogram(counts)
```

📌 **Resultados esperados:**  
✅ **El estado marcado tiene una probabilidad significativamente mayor de ser medido**.  
✅ **En un sistema clásico, hubiéramos necesitado muchas más iteraciones para encontrar el estado deseado.**
