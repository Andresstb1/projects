## **1Ô∏è‚É£ Teor√≠a de la Teleportaci√≥n Cu√°ntica**  

### **üîπ ¬øQu√© es la teleportaci√≥n cu√°ntica?**  
La **teleportaci√≥n cu√°ntica** permite transferir el **estado de un qubit** a otro **sin que exista una conexi√≥n f√≠sica directa entre ellos**. Esta t√©cnica se basa en **el entrelazamiento cu√°ntico** y el uso de **bits cl√°sicos para la transmisi√≥n de informaci√≥n**.

üí° **Importante:**  
- No se trata de mover f√≠sicamente un qubit de un lugar a otro.  
- La informaci√≥n cu√°ntica se transfiere a trav√©s de **qubits entrelazados y comunicaci√≥n cl√°sica**.  
- Es clave para **redes cu√°nticas, correcci√≥n de errores y comunicaciones cu√°nticas seguras**.

---

### **üîπ Concepto Matem√°tico**
Supongamos que queremos **transferir el estado de un qubit $A$** a otro qubit **$C$** usando un **qubit auxiliar $B$**.  

1Ô∏è‚É£ **Estado inicial del qubit $A$** (desconocido):  
$$
|\psi\rangle_A = \alpha |0\rangle + \beta |1\rangle
$$
2Ô∏è‚É£ **Entrelazamiento entre los qubits $B$ y $C$**:  
$$
|\Phi^+\rangle_{BC} = \frac{1}{\sqrt{2}} (|00\rangle_{BC} + |11\rangle_{BC})
$$
3Ô∏è‚É£ **El sistema total es la combinaci√≥n de ambos**:  
$$
|\psi\rangle_A \otimes |\Phi^+\rangle_{BC}
$$
4Ô∏è‚É£ **Medici√≥n conjunta en $A$ y $B$** y transmisi√≥n de bits cl√°sicos.  
5Ô∏è‚É£ **Correcci√≥n en $C$ mediante operaciones condicionales**, recuperando $|\psi\rangle$ en **$C$**.

---

### **üîπ Pasos del Protocolo de Teleportaci√≥n**  

1Ô∏è‚É£ **Generar un par de qubits entrelazados** entre **B** y **C**.  
2Ô∏è‚É£ **Aplicar una medici√≥n Bell a los qubits A y B**.  
3Ô∏è‚É£ **Enviar el resultado de la medici√≥n a trav√©s de un canal cl√°sico**.  
4Ô∏è‚É£ **Aplicar operaciones cu√°nticas en C basadas en la medici√≥n** para reconstruir el estado original.

---

## **2Ô∏è‚É£ Implementaci√≥n en Qiskit**

### **üìå Explicaci√≥n del Circuito**
üìå Para implementar la teleportaci√≥n cu√°ntica en **Qiskit**, seguimos estos pasos:

1Ô∏è‚É£ **Creamos 3 qubits**:  
   - **Qubit A:** Contiene el estado a teletransportar.  
   - **Qubit B y Qubit C:** Est√°n entrelazados para permitir la teleportaci√≥n.  

2Ô∏è‚É£ **Preparamos el entrelazamiento entre los qubits B y C** usando una **puerta Hadamard (H) y CNOT**.  

3Ô∏è‚É£ **Realizamos una medici√≥n Bell** en los qubits A y B.  

4Ô∏è‚É£ **Usamos los bits cl√°sicos para corregir el estado en C** con **puertas X y Z condicionales**.

---

### **üìå C√≥digo en Qiskit**

```python
# Importar librer√≠as necesarias
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram, array_to_latex
from qiskit.quantum_info import Statevector

# Simulador cu√°ntico
simulator = Aer.get_backend('aer_simulator')

# Crear registros cu√°nticos y cl√°sicos
qr = QuantumRegister(3, 'q')  # 3 qubits
cr = ClassicalRegister(2, 'c')  # 2 bits cl√°sicos
qc = QuantumCircuit(qr, cr)

# -------------------------------
# üîπ Paso 1: Preparar el estado a teletransportar
# -------------------------------
qc.initialize([0.6, 0.8], 0)  # Estado arbitrario en qubit A

# -------------------------------
# üîπ Paso 2: Crear el entrelazamiento entre B y C
# -------------------------------
qc.h(1)  # Hadamard en el qubit B
qc.cx(1, 2)  # CNOT entre qubit B y C

# -------------------------------
# üîπ Paso 3: Medici√≥n Bell en qubits A y B
# -------------------------------
qc.cx(0, 1)  # CNOT de A a B
qc.h(0)  # Hadamard en A

qc.measure(0, 0)  # Medir A
qc.measure(1, 1)  # Medir B

# -------------------------------
# üîπ Paso 4: Correcci√≥n del estado en qubit C
# -------------------------------
qc.x(2).c_if(cr, 1)  # Aplicar X si el bit cl√°sico 1 es 1
qc.z(2).c_if(cr, 2)  # Aplicar Z si el bit cl√°sico 0 es 1

# Dibujar el circuito
qc.draw('mpl')
```

## **3Ô∏è‚É£ Ejecuci√≥n y An√°lisis de Resultados**

### **üìå Ejecuci√≥n en el Simulador**

```python
# Transpilar el circuito y ejecutarlo en el simulador
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)

# Obtener resultados de medici√≥n
result = job.result()
counts = result.get_counts(qc)

# Mostrar histograma de resultados
plot_histogram(counts)
```

### **üìå Estado Final de Qubit C (Verificaci√≥n)**

```python
# -------------------------------
# üîπ An√°lisis del Estado Antes de la Medici√≥n
# -------------------------------
# Creamos una versi√≥n sin mediciones para calcular el estado cu√°ntico
qc_sin_medicion = qc.remove_final_measurements(inplace=False)

# Obtener el estado cu√°ntico antes de la medici√≥n
estado_final = Statevector(qc_sin_medicion)
print("\nEstado final antes de la medici√≥n:")
display(array_to_latex(estado_final.data, prefix="|œà‚ü© = "))

# -------------------------------
# üîπ Ejecuci√≥n en el Simulador
# -------------------------------
# Transpilar el circuito y ejecutarlo en el simulador
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)

# Obtener resultados de medici√≥n
result = job.result()
counts = result.get_counts(qc)

# Mostrar histograma de resultados
plot_histogram(counts)
```

# **üîπ Algoritmo de Deutsch-Jozsa**
El **Algoritmo de Deutsch-Jozsa** fue uno de los primeros en demostrar una **ventaja cu√°ntica clara** sobre los algoritmos cl√°sicos.

### **üìå ¬øCu√°l es el Problema?**
Tenemos una **funci√≥n $f(x)$** que acepta un **bit de entrada (x)** y devuelve **0 o 1**.  

Sabemos que la funci√≥n es:  
1Ô∏è‚É£ **Constante:** Siempre devuelve el mismo valor para cualquier entrada.  
2Ô∏è‚É£ **Balanceada:** Devuelve 0 la mitad de las veces y 1 la otra mitad.  

**Objetivo:**  
üìå **Determinar si la funci√≥n es constante o balanceada con solo UNA evaluaci√≥n**.  

### **üìå Implementaci√≥n en Qiskit**


```python
# Importar librer√≠as necesarias con Qiskit 1.0+
from qiskit.circuit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram

# Crear el circuito cu√°ntico para Deutsch-Jozsa con 2 qubits
qc = QuantumCircuit(2, 1)

# Inicializar en superposici√≥n usando Hadamard
qc.h([0, 1])

# Aplicar una puerta CNOT con el primer qubit como control y el segundo como objetivo
qc.cx(0, 1)

# Aplicar Hadamard nuevamente
qc.h(0)

# Medir el primer qubit
qc.measure(0, 0)

# Dibujar el circuito
qc.draw('mpl')

# Simulador cu√°ntico
simulator = Aer.get_backend('aer_simulator')
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)
result = job.result()

# Obtener los resultados
counts = result.get_counts(qc)
plot_histogram(counts)
```

üìå **Resultados esperados:**  
‚úÖ **Si la funci√≥n es constante**, siempre obtenemos **"0"**.  
‚úÖ **Si la funci√≥n es balanceada**, obtenemos **"1"**.  

‚úÖ **Supera a cualquier algoritmo cl√°sico, que necesitar√≠a m√∫ltiples evaluaciones para verificar todas las entradas.**

# **üîπ Algoritmo de Grover**
El **Algoritmo de Grover** resuelve el problema de **b√∫squeda en bases de datos desordenadas** con **una mejora cuadr√°tica** sobre los algoritmos cl√°sicos.

### **üìå ¬øCu√°l es el Problema?**
Dada una **base de datos de $N$ elementos**, queremos encontrar un **elemento marcado**.  

üìå **En una computadora cl√°sica, esto tomar√≠a $\mathcal{O}(N)$ intentos** en promedio.  
üìå **Grover lo reduce a $\mathcal{O}(\sqrt{N})$ intentos** usando **amplificaci√≥n de amplitud**.  

---

### **üìå Implementaci√≥n en Qiskit**

```python
# Importar librer√≠as necesarias con Qiskit 1.0+
from qiskit.circuit import QuantumCircuit
from qiskit_aer import Aer  # ‚úÖ Importaci√≥n corregida para Qiskit 1.0+
from qiskit import transpile
from qiskit.visualization import plot_histogram

# Crear un circuito con 2 qubits y 2 bits cl√°sicos
qc = QuantumCircuit(2, 2)

# Inicializar en superposici√≥n
qc.h([0, 1])

# Aplicar una puerta X en el segundo qubit (marcar un estado)
qc.x(1)

# Aplicar una compuerta Z controlada para amplificar el estado marcado
qc.cz(0, 1)

# Aplicar Hadamard nuevamente
qc.h([0, 1])

# Medir ambos qubits
qc.measure([0, 1], [0, 1])

# Dibujar el circuito
qc.draw('mpl')

# Ejecutar el circuito en el simulador cu√°ntico
simulator = Aer.get_backend('aer_simulator')
transpiled_qc = transpile(qc, simulator)
job = simulator.run(transpiled_qc, shots=1024)
result = job.result()

# Obtener y visualizar los resultados
counts = result.get_counts(qc)
plot_histogram(counts)
```

üìå **Resultados esperados:**  
‚úÖ **El estado marcado tiene una probabilidad significativamente mayor de ser medido**.  
‚úÖ **En un sistema cl√°sico, hubi√©ramos necesitado muchas m√°s iteraciones para encontrar el estado deseado.**

---

# **üìå Conclusi√≥n**
‚úÖ **El Algoritmo de Deutsch-Jozsa demuestra c√≥mo un problema que requer√≠a m√∫ltiples evaluaciones cl√°sicas se resuelve con una sola evaluaci√≥n cu√°ntica.**  
‚úÖ **El Algoritmo de Grover muestra c√≥mo la interferencia cu√°ntica puede amplificar las soluciones correctas, reduciendo el tiempo de b√∫squeda.**  
‚úÖ **Estos algoritmos representan los primeros pasos hacia la ventaja cu√°ntica en problemas pr√°cticos.**
