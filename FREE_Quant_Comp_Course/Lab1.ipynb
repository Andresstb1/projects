### **¿Qué es un qubit y cómo se diferencia de un bit clásico?**

#### **1. El bit clásico vs. el qubit**
En la computación tradicional, la unidad más básica de información es el **bit**, que puede tomar uno de dos valores: **0 o 1**. En cambio, en la computación cuántica, la unidad de información es el **qubit** (abreviatura de *quantum bit*), que no solo puede estar en los estados **$|0\rangle$** y **$|1\rangle$**, sino también en cualquier **combinación lineal** de ambos, lo que se conoce como **superposición**.

| **Característica**  | **Bit clásico** | **Qubit** |
|------------------|--------------|--------|
| Estado posible | $0$ o $1$ | Cualquier combinación de $|0\rangle$ y $|1\rangle$ |
| Almacenamiento | Determinista | Probabilístico |
| Representación matemática | $0$ o $1$ | Estado cuántico con amplitudes de probabilidad |
| Medición | Devuelve $0$ o $1$ | Colapsa a $|0\rangle$ o $|1\rangle$ con cierta probabilidad |

---

#### **2. Representación matemática del qubit**
Un qubit se representa como un **vector en un espacio de Hilbert** de dos dimensiones:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

donde:

- $\alpha$ y $\beta$ son **números complejos** llamados **amplitudes de probabilidad**.
- La probabilidad de obtener el resultado $|0\rangle$ al medir el qubit es $|\alpha|^2$.
- La probabilidad de obtener el resultado $|1\rangle$ al medir el qubit es $|\beta|^2$.
- La norma del estado debe estar **normalizada**, es decir:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

---

#### **Ejemplo: Estado en superposición**
Un qubit puede estar en un estado como:

$$
|\psi\rangle = \frac{1}{\sqrt{2}} |0\rangle + \frac{1}{\sqrt{2}} |1\rangle
$$

En este caso, si medimos el qubit, tenemos **50% de probabilidad** de obtener **$|0\rangle$** y **50% de probabilidad** de obtener **$|1\rangle$**.

## **2. Importar las dependencias necesarias**
Una vez instalado **Qiskit**, debemos importar los módulos clave para la creación de circuitos y visualización de qubits. En **Qiskit 1.0**, la estructura de los módulos ha cambiado, por lo que debemos asegurarnos de utilizar las nuevas importaciones:

````python
# Importar módulos principales de Qiskit
from qiskit.circuit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_bloch_vector, plot_histogram

# Importar librerías adicionales
import numpy as np
import matplotlib.pyplot as plt
````

**Explicación de las importaciones:**

- `plot_histogram()`: Permite visualizar resultados de medición en forma de histograma.
- `numpy` y `matplotlib.pyplot`: Son librerías auxiliares para cálculos y gráficos.

### **3. Creando y visualizando un qubit en estado inicial**
En **Qiskit**, cuando creamos un qubit, su estado inicial es **$|0\rangle$**, es decir, se encuentra en el polo norte de la **esfera de Bloch**. Veamos cómo programarlo:

````python
# Crear un circuito cuántico con 1 qubit
qc = QuantumCircuit(1)

# Dibujar el circuito
qc.draw('mpl')
````

Este código genera un circuito muy simple con un solo qubit. Como no hemos aplicado ninguna transformación, el qubit permanece en el estado **$|0\rangle$**.

Ahora visualicemos su estado en la **esfera de Bloch**:


````python
# Representación del estado inicial del qubit en la esfera de Bloch
plot_bloch_vector([0, 0, 1], title="Estado inicial |0⟩ en la esfera de Bloch")
````

**🔹 Explicación de la visualización en la esfera de Bloch:**
- La esfera de Bloch es una representación geométrica del estado del qubit.
- En su estado inicial **$|0\rangle$**, el qubit se encuentra en la parte superior de la esfera.
- En coordenadas cartesianas, esto corresponde a **(x=0, y=0, z=1)**.

---

### **4. Cambiando el estado del qubit a $|1\rangle$**
Podemos aplicar una **puerta X** (*NOT cuántico*) para cambiar el estado del qubit de **$|0\rangle$** a **$|1\rangle$**:


````python
# Aplicar una puerta X al qubit
qc.x(0)

# Dibujar el circuito actualizado
qc.draw('mpl')
````

````python
# Representación del estado |1⟩ en la esfera de Bloch
plot_bloch_vector([0, 0, -1], title="Estado |1⟩ en la esfera de Bloch")
````

📌 **Observación**:
- Ahora el qubit está en el polo sur de la esfera de Bloch, correspondiente al estado **$|1\rangle$**.
- En coordenadas cartesianas: **(x=0, y=0, z=-1)**.

### **5. Medición de un qubit en Qiskit**
Hasta ahora, hemos creado un qubit, lo hemos visualizado en la **esfera de Bloch** y aplicamos una **puerta X** para cambiar su estado. Ahora aprenderemos cómo **medir un qubit** y analizar los resultados obtenidos.

En computación cuántica, la **medición** de un qubit **colapsa su estado cuántico** a uno de los estados clásicos: **$|0\rangle$ o $|1\rangle$**. La probabilidad de obtener cada resultado está dada por el cuadrado del módulo de sus **amplitudes de probabilidad**.

Por ejemplo, si un qubit está en un estado general:
$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$
- La probabilidad de medir **$|0\rangle$** es **$|\alpha|^2$**.
- La probabilidad de medir **$|1\rangle$** es **$|\beta|^2$**.
- Estas probabilidades siempre suman **1**, es decir, **$|\alpha|^2 + |\beta|^2 = 1$**.

---

### **6. Implementación de la medición en Qiskit**
En Qiskit, para medir un qubit debemos **agregar una operación de medición** en nuestro circuito y luego **ejecutarlo en un simulador**.



````python
# Crear un circuito cuántico con 1 qubit y 1 bit clásico para almacenar la medición
qc = QuantumCircuit(1, 1)

# Aplicamos una puerta Hadamard (H) para crear un estado en superposición
qc.h(0)

# Agregar la medición
qc.measure(0, 0)

# Dibujar el circuito
qc.draw('mpl')
````

#### **Explicación del código:**
1. **Creamos un circuito con 1 qubit y 1 bit clásico.**
2. **Aplicamos una puerta Hadamard (H)** al qubit. Esto lo coloca en un **estado de superposición**:
   $$
   |\psi\rangle = \frac{1}{\sqrt{2}} |0\rangle + \frac{1}{\sqrt{2}} |1\rangle
   $$
   Ahora el qubit tiene **50% de probabilidad** de colapsar en **$|0\rangle$** y **50% de probabilidad** de colapsar en **$|1\rangle$** cuando lo medimos.
3. **Agregamos una operación de medición** usando `qc.measure(0, 0)`, lo que mide el qubit 0 y almacena el resultado en el bit clásico 0.

---

### **7. Ejecución del circuito y visualización de los resultados**
Para ejecutar nuestro circuito, utilizaremos un **simulador cuántico** de Qiskit.


````python
# Seleccionar el simulador cuántico de Aer
simulator = Aer.get_backend('aer_simulator')

# Transpilar el circuito para optimizarlo
transpiled_qc = transpile(qc, simulator)

# Ejecutar el circuito en el simulador
job = simulator.run(transpiled_qc, shots=1024)

# Obtener los resultados de la medición
result = job.result()
counts = result.get_counts(qc)

# Mostrar los resultados en forma de histograma
plot_histogram(counts)
````

### **8. Análisis de los resultados**
Cuando ejecutamos este código, obtenemos un **histograma de resultados**, que representa cuántas veces se midió **$|0\rangle$** y cuántas veces se midió **$|1\rangle$** en **1024 repeticiones (shots)**.

Dado que aplicamos una puerta Hadamard al qubit, esperamos que el qubit esté en una **superposición equilibrada**, lo que significa que aproximadamente **la mitad de las veces obtenemos $|0\rangle$ y la otra mitad obtenemos $|1\rangle$**.

Si ejecutamos el circuito, obtenemos un **resultado similar a este:**
```
 {'0': 510, '1': 514}
```
Lo que indica que:
- Se obtuvo **$|0\rangle$** en **aproximadamente 50% de las mediciones**.
- Se obtuvo **$|1\rangle$** en **aproximadamente 50% de las mediciones**.

Esto es consistente con la teoría, ya que el estado en superposición tiene una **probabilidad del 50%** para cada resultado.
