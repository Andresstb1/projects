### **驴Qu茅 es un qubit y c贸mo se diferencia de un bit cl谩sico?**

#### **1. El bit cl谩sico vs. el qubit**
En la computaci贸n tradicional, la unidad m谩s b谩sica de informaci贸n es el **bit**, que puede tomar uno de dos valores: **0 o 1**. En cambio, en la computaci贸n cu谩ntica, la unidad de informaci贸n es el **qubit** (abreviatura de *quantum bit*), que no solo puede estar en los estados **$|0\rangle$** y **$|1\rangle$**, sino tambi茅n en cualquier **combinaci贸n lineal** de ambos, lo que se conoce como **superposici贸n**.

| **Caracter铆stica**  | **Bit cl谩sico** | **Qubit** |
|------------------|--------------|--------|
| Estado posible | $0$ o $1$ | Cualquier combinaci贸n de $|0\rangle$ y $|1\rangle$ |
| Almacenamiento | Determinista | Probabil铆stico |
| Representaci贸n matem谩tica | $0$ o $1$ | Estado cu谩ntico con amplitudes de probabilidad |
| Medici贸n | Devuelve $0$ o $1$ | Colapsa a $|0\rangle$ o $|1\rangle$ con cierta probabilidad |

---

#### **2. Representaci贸n matem谩tica del qubit**
Un qubit se representa como un **vector en un espacio de Hilbert** de dos dimensiones:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

donde:

- $\alpha$ y $\beta$ son **n煤meros complejos** llamados **amplitudes de probabilidad**.
- La probabilidad de obtener el resultado $|0\rangle$ al medir el qubit es $|\alpha|^2$.
- La probabilidad de obtener el resultado $|1\rangle$ al medir el qubit es $|\beta|^2$.
- La norma del estado debe estar **normalizada**, es decir:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

---

#### **Ejemplo: Estado en superposici贸n**
Un qubit puede estar en un estado como:

$$
|\psi\rangle = \frac{1}{\sqrt{2}} |0\rangle + \frac{1}{\sqrt{2}} |1\rangle
$$

En este caso, si medimos el qubit, tenemos **50% de probabilidad** de obtener **$|0\rangle$** y **50% de probabilidad** de obtener **$|1\rangle$**.

### **3. Creando y visualizando un qubit en estado inicial**
En **Qiskit**, cuando creamos un qubit, su estado inicial es **$|0\rangle$**, es decir, se encuentra en el polo norte de la **esfera de Bloch**. Veamos c贸mo programarlo:

````python
# Crear un circuito cu谩ntico con 1 qubit
qc = QuantumCircuit(1)

# Dibujar el circuito
qc.draw('mpl')
````

Este c贸digo genera un circuito muy simple con un solo qubit. Como no hemos aplicado ninguna transformaci贸n, el qubit permanece en el estado **$|0\rangle$**.

Ahora visualicemos su estado en la **esfera de Bloch**:


````python
# Representaci贸n del estado inicial del qubit en la esfera de Bloch
plot_bloch_vector([0, 0, 1], title="Estado inicial |0 en la esfera de Bloch")
````

** Explicaci贸n de la visualizaci贸n en la esfera de Bloch:**

### **4. Cambiando el estado del qubit a $|1\rangle$**
Podemos aplicar una **puerta X** (*NOT cu谩ntico*) para cambiar el estado del qubit de **$|0\rangle$** a **$|1\rangle$**:

### **5. Medici贸n de un qubit en Qiskit**
Hasta ahora, hemos creado un qubit, lo hemos visualizado en la **esfera de Bloch** y aplicamos una **puerta X** para cambiar su estado. Ahora aprenderemos c贸mo **medir un qubit** y analizar los resultados obtenidos.

En computaci贸n cu谩ntica, la **medici贸n** de un qubit **colapsa su estado cu谩ntico** a uno de los estados cl谩sicos: **$|0\rangle$ o $|1\rangle$**. La probabilidad de obtener cada resultado est谩 dada por el cuadrado del m贸dulo de sus **amplitudes de probabilidad**.

Por ejemplo, si un qubit est谩 en un estado general:
$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$
- La probabilidad de medir **$|0\rangle$** es **$|\alpha|^2$**.
- La probabilidad de medir **$|1\rangle$** es **$|\beta|^2$**.
- Estas probabilidades siempre suman **1**, es decir, **$|\alpha|^2 + |\beta|^2 = 1$**.

---

### **6. Implementaci贸n de la medici贸n en Qiskit**
En Qiskit, para medir un qubit debemos **agregar una operaci贸n de medici贸n** en nuestro circuito y luego **ejecutarlo en un simulador**.



````python
# Crear un circuito cu谩ntico con 1 qubit y 1 bit cl谩sico para almacenar la medici贸n
qc = QuantumCircuit(1, 1)

# Aplicamos una puerta Hadamard (H) para crear un estado en superposici贸n
qc.h(0)

# Agregar la medici贸n
qc.measure(0, 0)

# Dibujar el circuito
qc.draw('mpl')
````

#### **Explicaci贸n del c贸digo:**
1. **Creamos un circuito con 1 qubit y 1 bit cl谩sico.**
2. **Aplicamos una puerta Hadamard (H)** al qubit. Esto lo coloca en un **estado de superposici贸n**:
   $$
   |\psi\rangle = \frac{1}{\sqrt{2}} |0\rangle + \frac{1}{\sqrt{2}} |1\rangle
   $$
   Ahora el qubit tiene **50% de probabilidad** de colapsar en **$|0\rangle$** y **50% de probabilidad** de colapsar en **$|1\rangle$** cuando lo medimos.
3. **Agregamos una operaci贸n de medici贸n** usando `qc.measure(0, 0)`, lo que mide el qubit 0 y almacena el resultado en el bit cl谩sico 0.

---

### **7. Ejecuci贸n del circuito y visualizaci贸n de los resultados**
Para ejecutar nuestro circuito, utilizaremos un **simulador cu谩ntico** de Qiskit.


````python
# Seleccionar el simulador cu谩ntico de Aer
simulator = Aer.get_backend('aer_simulator')

# Transpilar el circuito para optimizarlo
transpiled_qc = transpile(qc, simulator)

# Ejecutar el circuito en el simulador
job = simulator.run(transpiled_qc, shots=1024)

# Obtener los resultados de la medici贸n
result = job.result()
counts = result.get_counts(qc)

# Mostrar los resultados en forma de histograma
plot_histogram(counts)
````

### **8. An谩lisis de los resultados**
Cuando ejecutamos este c贸digo, obtenemos un **histograma de resultados**, que representa cu谩ntas veces se midi贸 **$|0\rangle$** y cu谩ntas veces se midi贸 **$|1\rangle$** en **1024 repeticiones (shots)**.

Dado que aplicamos una puerta Hadamard al qubit, esperamos que el qubit est茅 en una **superposici贸n equilibrada**, lo que significa que aproximadamente **la mitad de las veces obtenemos $|0\rangle$ y la otra mitad obtenemos $|1\rangle$**.

Si ejecutamos el circuito, obtenemos un **resultado similar a este:**
```
 {'0': 510, '1': 514}
```
Lo que indica que:
- Se obtuvo **$|0\rangle$** en **aproximadamente 50% de las mediciones**.
- Se obtuvo **$|1\rangle$** en **aproximadamente 50% de las mediciones**.

Esto es consistente con la teor铆a, ya que el estado en superposici贸n tiene una **probabilidad del 50%** para cada resultado.
