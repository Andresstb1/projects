<h4 style="font-size:18px; margin-bottom:10px; color:#2c3e50;">3.1 Classical Bit vs. Quantum Bit (Qubit)</h4>
<p>
  In traditional computing, the basic unit of information is the <strong>bit</strong>, which can take one of two values: <strong>0</strong> or <strong>1</strong>.  
  In quantum computing, the equivalent unit is the <strong>qubit</strong> (short for <em>quantum bit</em>), which can exist not only in the states 
  <strong>|0⟩</strong> and <strong>|1⟩</strong>, but also in any <em>linear combination</em> of both — a phenomenon known as <strong>superposition</strong>.
</p>

<h4 style="font-size:18px; margin-bottom:10px; color:#2c3e50;">3.2 Mathematical Representation of a Qubit and Measurements</h4>

<p>
  In quantum mechanics, the state of a system — such as a single qubit — is described using the <strong>Dirac notation</strong>, also known as the <em>bra-ket notation</em>.  
  This formalism provides a compact and elegant way to represent <strong>vectors</strong> and <strong>linear operations</strong> in complex vector spaces, known as <em>Hilbert spaces</em>.
</p>

<p>
  A <strong>ket</strong> (written as \( |\psi\rangle \)) represents a <em>column vector</em> in this space, while its corresponding <strong>bra</strong> (written as \( \langle\psi| \)) represents the <em>row vector</em> obtained by taking the complex conjugate transpose of the ket.  
  Formally:
</p>

<p style="text-align:center; font-size:17px;">
  \( |\psi\rangle =
  \begin{bmatrix}
    \alpha \\
    \beta
  \end{bmatrix},
  \quad
  \langle\psi| =
  \begin{bmatrix}
    \alpha^* & \beta^*
  \end{bmatrix}
  \)
</p>

<p>
  In this framework, a single qubit is represented as a <strong>vector in a two-dimensional complex Hilbert space</strong>:
</p>

<p style="text-align:center; font-size:17px;">
  \( |\psi\rangle = \alpha |0\rangle + \beta |1\rangle \)
</p>

<p>
  Here:
</p>
<ul>
  <li><strong>|0⟩</strong> and <strong>|1⟩</strong> form the <em>computational basis</em> of the space — two orthonormal kets that span all possible qubit states.</li>
  <li>\( \alpha \) and \( \beta \) are <strong>complex probability amplitudes</strong> associated with each basis state.</li>
</ul>

<p>
  In matrix form, these basis vectors can be explicitly written as:
</p>

<p style="text-align:center; font-size:17px;">
  \( |0\rangle =
  \begin{bmatrix}
    1 \\
    0
  \end{bmatrix},
  \quad
  |1\rangle =
  \begin{bmatrix}
    0 \\
    1
  \end{bmatrix}
  \)
</p>

<p>
  The coefficients \( \alpha \) and \( \beta \) encode both the <em>likelihood</em> and the <em>phase relationship</em> of the basis states.  
  The measurable probabilities are given by the <strong>squared magnitudes</strong> of these amplitudes:
</p>

<ul>
  <li>Probability of obtaining <strong>|0⟩</strong>: \( |\alpha|^2 \)</li>
  <li>Probability of obtaining <strong>|1⟩</strong>: \( |\beta|^2 \)</li>
</ul>

<p>
  Since the total probability of finding the system in one of its basis states must be 1, the state must satisfy the <strong>normalization condition</strong>:
</p>

<p style="text-align:center; font-size:17px;">
  \( |\alpha|^2 + |\beta|^2 = 1 \)
</p>

<p>
  In the Dirac formalism, the <strong>inner product</strong> (or scalar product) between two states is written as:
</p>

<p style="text-align:center; font-size:17px;">
  \( \langle\phi|\psi\rangle = \alpha_\phi^* \alpha_\psi + \beta_\phi^* \beta_\psi \)
</p>

<p>
  which yields a complex number that measures their <em>overlap</em> — a key concept for determining probabilities and quantum interference.
</p>

<p>
  When a measurement is performed, the qubit’s state <strong>collapses</strong> to one of the basis states (|0⟩ or |1⟩),  
  and all superposition information encoded in \( \alpha \) and \( \beta \) is lost.  
  Each measurement outcome occurs with probability proportional to the square of the corresponding amplitude.
</p>

<p>
  Thus, the mathematical structure of quantum mechanics captures two complementary aspects:
</p>
<ul>
  <li>The <strong>continuous nature</strong> of quantum states, represented by complex vectors in Hilbert space.</li>
  <li>The <strong>discrete nature</strong> of measurement outcomes, emerging from the probabilistic collapse of those vectors onto basis states.</li>
</ul>

<p>
  This is what gives rise to the fundamental distinction between classical bits and quantum bits,  
  and to the inherently probabilistic character of quantum computation.
</p>



<h4 style="font-size:18px; margin-bottom:10px; color:#2c3e50;">3.3 The Bloch Sphere Representation</h4>

<p>
  The <strong>Bloch sphere</strong> provides a powerful geometric picture of the qubit state as a point on the surface of a unit sphere in 3D space.  
  It translates the complex-valued amplitudes \( \alpha \) and \( \beta \) into real angular coordinates, offering a more intuitive visualization of quantum behavior.
</p>

<p>
  Any qubit state can be rewritten using two real parameters — <strong>θ</strong> (theta) and <strong>φ</strong> (phi) — as:
</p>

<p style="text-align:center; font-size:17px;">
  \( |\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\right)|1\rangle \)
</p>

<p>
  This form highlights the geometric correspondence between the qubit’s algebraic description and its position on the sphere:
</p>
<ul>
  <li><strong>θ</strong> determines the <em>polar angle</em> (latitude), measuring how far the state is tilted from the north pole (|0⟩).</li>
  <li><strong>φ</strong> determines the <em>azimuthal angle</em> (longitude), encoding the <strong>relative phase</strong> between |0⟩ and |1⟩.</li>
</ul>

<p>
  The north pole (<strong>θ = 0</strong>) represents <strong>|0⟩</strong>, the south pole (<strong>θ = π</strong>) represents <strong>|1⟩</strong>,  
  and all other points on the surface correspond to <em>superpositions</em> of these basis states.
</p>

<p>
  Importantly, this representation arises naturally from the normalization condition \( |\alpha|^2 + |\beta|^2 = 1 \):  
  the qubit’s state vector always has unit length, just like any point on a sphere.
</p>

<p>
  For example, the <strong>Hadamard gate</strong> transforms <strong>|0⟩</strong> into:
</p>

<p style="text-align:center; font-size:17px;">
  \( |+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \)
</p>

<p>
  which lies on the <strong>x-axis</strong> of the Bloch sphere — midway between |0⟩ and |1⟩.  
  Quantum gates, therefore, can be interpreted geometrically as <strong>rotations of the state vector</strong> around the sphere’s axes.
</p>

<p>
  The Bloch sphere bridges two complementary perspectives:
</p>
<ul>
  <li>The <strong>algebraic</strong> view — where qubits are complex vectors defined by amplitudes (α, β).</li>
  <li>The <strong>geometric</strong> view — where the same qubits correspond to points on a sphere parameterized by (θ, φ).</li>
</ul>

<p>
  Together, these perspectives reveal the continuous and complex nature of qubits before measurement,  
  and how discrete outcomes arise when those states collapse upon observation.
</p>


<h4 style="font-size:18px; margin-bottom:10px; color:#2c3e50;">3.4 The Role of Phase in Quantum States</h4>

<p>
  In the Bloch sphere formalism, the angle <strong>φ</strong> encodes the <em>relative phase</em> between the components of a superposition.  
  This phase is crucial because it determines <strong>how qubits interfere</strong> with each other during quantum operations — a property with no classical analog.
</p>

<p>
  Consider two states:
</p>

<p style="text-align:center; font-size:17px;">
  \( |+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \quad \text{and} \quad |-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) \)
</p>

<p>
  Both have identical measurement probabilities (50% for |0⟩ and 50% for |1⟩),  
  yet they differ by a <strong>relative phase of π</strong>.  
  This difference becomes meaningful only when these states interact with other qubits or gates — for instance, in interference experiments or quantum algorithms.
</p>

<p>
  When a qubit is measured, this phase information disappears, since measurement projects the state onto one of the basis vectors (|0⟩ or |1⟩).  
  However, <em>before measurement</em>, relative phase governs the evolution of quantum systems and determines the results of subsequent operations.
</p>

<p>
  In summary, while global phase (a common factor \( e^{i\gamma} \) applied to the entire state) has no physical effect,  
  <strong>relative phase</strong> — the difference in phase between the amplitudes of |0⟩ and |1⟩ — is fundamental to the power of quantum computation.
</p>

  <h3 style="margin-top:0.2;">5.1 Creating and Visualizing an Initial Qubit State</h3>

  <p>
    In <strong>Qiskit</strong>, when a qubit is created, it is automatically initialized in the state <strong>|0⟩</strong>.  
    This state represents a definite quantum configuration with no superposition or entanglement yet applied.
  </p>

  <h4><strong>Understanding the Bloch Sphere</strong></h4>
  <p>
    The <strong>Bloch sphere</strong> is a powerful geometric representation of the qubit’s state within a three-dimensional space.  
    Every point on the surface of the sphere corresponds to a valid pure quantum state.
  </p>
  <ul>
    <li>The north and south poles correspond to the basis states <strong>|0⟩</strong> and <strong>|1⟩</strong>, respectively.</li>
    <li>Points on the surface between them represent <em>superpositions</em> of these two basis states.</li>
    <li>Changes to the qubit’s state through quantum gates can be visualized as rotations on this sphere.</li>
  </ul>

  <h4><strong>The Position of |0⟩ on the Bloch Sphere</strong></h4>
  <p>
    In its initial state <strong>|0⟩</strong>, the qubit is located at the <em>north pole</em> of the Bloch sphere.  
    In Cartesian coordinates, this corresponds to the point <strong>(x = 0, y = 0, z = 1)</strong>. Let's use Qiskit to represent this initial state as a cicuit and in the Bloch Sphere:
  </p>
</div>

```python
# --- Creating and Visualizing an Initial Qubit State ---

from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_vector

# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)

# Draw the circuit
qc.draw('mpl')
```

<div style="background-color: #f4f4f6; border-radius: 8px; padding: 15px 20px; font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.6;">
<p>
    This circuit defined a single qubit without any gate operations. Since no transformations are performed, the qubit remains in its initial state <strong>|0⟩</strong>.
  </p>
</div>

```python
# Visualize the initial qubit state on the Bloch sphere
# The |0⟩ state corresponds to the vector (x=0, y=0, z=1)
plot_bloch_vector([0, 0, 1], title="Initial |0⟩ State on the Bloch Sphere")
```

<div style="background-color: #f4f4f6; border-radius: 8px; padding: 15px 20px; font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.6;">
  <h3 style="margin-top:0.2;">5.2 Manipulating the Qubit State</h3>

  <p>
    So far, we have seen that a newly created qubit in <strong>Qiskit</strong> starts in the state <strong>|0⟩</strong>, located at the north pole of the <strong>Bloch sphere</strong>.  
    However, one of the most fascinating aspects of quantum computing is the ability to <strong>manipulate</strong> this state.
  </p>

  <p>
    In quantum circuits, this manipulation is achieved through <strong>quantum gates</strong> — operations that change the orientation of the qubit on the Bloch sphere.  
    For now, you don’t need to understand how these gates work in detail (you will learn about the quantum gates in the next labs).  
    Simply imagine that they are capable of <em>rotating</em> or <em>transforming</em> the qubit’s state in specific ways.
  </p>

  <p>
    Let’s explore how the application of simple transformations affects the qubit’s position on the Bloch sphere:
  </p>
</div>


```python
# Apply a "NOT" gate (X) to flip the value of the qubit into 1
qc.x(0)

# Draw the updated circuit
qc.draw('mpl')
```

```python
# Representation of the 1 state on the Bloch Sphere
plot_bloch_vector([0, 0, -1], title="State 1 on the Bloch Sphere")
```

```python
# Reset the quantum cicuit to state 0
qc = QuantumCircuit(1)

# Apply a Hadamard gate (H) to place the qubit in superposition
qc.h(0)

# Draw the updated circuit
qc.draw('mpl')
```

```python
# Visualize the resulting state on the Bloch sphere
# The |+⟩ state corresponds to the vector (x=1, y=0, z=0)
plot_bloch_vector([1, 0, 0], title="Superposition State |+⟩ on the Bloch Sphere")
```

<div style="background-color: #f4f4f6; border-radius: 8px; padding: 15px 20px; font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.6;">
  <p>  
    When applied to the initial state <strong>|0⟩</strong>, the Hadamard gate created a new state where the qubit is in an <em>equal superposition</em> of <strong>|0⟩</strong> and <strong>|1⟩</strong>:
  </p>

  <p style="text-align:center; font-size:17px;">
    \( |\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \)
  </p>

  <p>
    This means that if we measure the qubit, there is a <strong>50% probability</strong> of obtaining <strong>|0⟩</strong> and a <strong>50% probability</strong> of obtaining <strong>|1⟩</strong>.  
    On the Bloch sphere, this state lies along the <strong>x-axis</strong>.
  </p>
</div>

<div style="background-color: #f4f4f6; border-radius: 8px; padding: 15px 20px; font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.6;">
  <h3 style="margin-top:0.2;">5.3 Measuring Qubit States</h3>

  <p>
    Once a quantum circuit has been built, the next step is to <strong>measure</strong> the qubits.  
    Measurement in quantum computing is a process that converts a qubit’s quantum state into a <strong>classical bit</strong> — either <strong>0</strong> or <strong>1</strong>.  
    This action <em>collapses</em> the superposition of the qubit into one of its basis states, based on the probability amplitudes of the quantum state.
  </p>

  <p>
    In <strong>Qiskit</strong>, we measure qubits by connecting them to <strong>classical bits</strong>.  
    The result of the measurement is stored in these classical bits, which can then be read as standard binary data.  
    If the qubit was in a superposition, each measurement gives a random result according to the probabilities \( |\alpha|^2 \) and \( |\beta|^2 \).  
    Repeating the circuit multiple times allows us to observe these probabilities statistically.
  </p>

  <p>
    The following example measures a single qubit that has been placed in superposition using the Hadamard gate:
  </p>
</div>


```python
# --- Measuring a Single Qubit in Superposition ---

# Create a circuit with one qubit and one classical bit
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate to create a superposition state
qc.h(0)

# Measure the qubit and store the result in the classical bit
qc.measure(0, 0)

# Draw the circuit
qc.draw('mpl')
```

```python
# --- Running the Circuit and Visualizing the Results ---

# Use the Aer simulator
simulator = AerSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the circuit with multiple shots (repetitions)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()

# Get the measurement counts
counts = result.get_counts()
print("Measurement results:", counts)

# Plot a histogram of the outcomes
plot_histogram(counts, title="Measurement Outcomes for a Qubit in Superposition")
```

<div style="background-color: #f4f4f6; border-radius: 8px; padding: 15px 20px; font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.6;">
  <p>
    In this example, since the Hadamard gate creates an equal superposition of <strong>|0⟩</strong> and <strong>|1⟩</strong>,  
    we expect the results to be roughly <strong>50%</strong> zeros and <strong>50%</strong> ones after many measurements.  
    Each execution (or <em>shot</em>) produces one random outcome, but the overall distribution reveals the underlying quantum probabilities.
  </p>
  <p>
    Measurement thus acts as the <strong>bridge between quantum and classical information</strong> — it collapses a probabilistic state into a definite value that can be observed and processed by a classical computer.

#### Theoretical Answer:

In its initial configuration, a qubit in Qiskit starts in the state **|0⟩**. This is a **deterministic state**, meaning that every time it is measured, the result will always be 0. The state can be written in vector form as:

$$
|\psi\rangle = |0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

This vector indicates that the probability amplitude associated with **|0⟩** is 1, and with **|1⟩** is 0. Therefore, the probabilities of each possible measurement outcome are:

$$
P(0) = |\alpha|^2 = 1 \quad \text{and} \quad P(1) = |\beta|^2 = 0
$$

Since there is no uncertainty in the state, the histogram of measurement results will always show a single bar for the outcome **0**. This behavior illustrates a key concept of quantum mechanics — **state collapse** — but in this particular case, the qubit is already in one of the basis states, so no probabilistic behavior is observed.

<div style="background: #f4f4f6; border-left: 4px solid #2c3e50; padding: 15px; font-family: sans-serif;">
    <h3 style="margin-top:0;">Exercise 2: Use the Hadamard gate to place the qubit in superposition, then measure it multiple times and analyze the <strong>statistical distribution</strong> of the results.</h3>
</div>  

#### Code Answer:

```python
# --- Exercise 2: Measuring a Qubit in Superposition ---

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Create a quantum circuit with one qubit and one classical bit
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate to create a superposition of |0⟩ and |1⟩
qc.h(0)

# Measure the qubit and store the result in the classical bit
qc.measure(0, 0)

# Draw the circuit
qc.draw('mpl')

# Initialize the Aer simulator
simulator = AerSimulator()

# Transpile and execute the circuit with multiple shots
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()

# Retrieve measurement counts
counts = result.get_counts()

# Display the results
print("Measurement results:", counts)

# Plot a histogram of measurement outcomes
plot_histogram(counts, title="Measurement Results for a Qubit in Superposition (|+⟩ State)")
```

#### Theoretical Answer:

When a **Hadamard gate** is applied to a qubit initially in the state **|0⟩**, the resulting state becomes an **equal superposition** of the two basis states |0⟩ and |1⟩. Mathematically, this is expressed as:

$$
|\psi\rangle = H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
$$

This means that the qubit now has an **equal probability** of being measured as 0 or 1. The measurement process collapses the superposed state into one of these outcomes, but which one occurs in each experiment is inherently random.

If the same circuit is executed many times, each run (or **shot**) yields a different random result, but the **overall frequency** of outcomes will tend toward 50% for |0⟩ and 50% for |1⟩. This reflects the squared magnitudes of the corresponding amplitudes:

$$
P(0) = \left|\frac{1}{\sqrt{2}}\right|^2 = \frac{1}{2}, \quad
P(1) = \left|\frac{1}{\sqrt{2}}\right|^2 = \frac{1}{2}
$$

The histogram obtained from these measurements shows two bars of approximately equal height, illustrating one of the most fundamental ideas in quantum mechanics: **probabilistic measurement outcomes arising from deterministic state evolution**.

<div style="background: #f4f4f6; border-left: 4px solid #2c3e50; padding: 15px; font-family: sans-serif;">
    <h3 style="margin-top:0;">Exercise 3: <strong>Experiment</strong> with different numbers of shots (e.g., 100, 1000, 5000) and observe how the <strong>accuracy</strong> of the probability estimation changes.</h3>
</div>  

#### Code Answer:

```python
# --- Exercise 3: Effect of the Number of Shots on Measurement Statistics ---

# Ensures that matplotlib displays plots directly inside the Jupyter notebook
%matplotlib inline  

# Create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)
qc.h(0)           # Apply a Hadamard gate to create a superposition state
qc.measure(0, 0)  # Measure the qubit and store the result in the classical bit

# Initialize the simulator backend
simulator = AerSimulator()

# Transpile the circuit so it can be executed by the chosen backend
compiled_circuit = transpile(qc, simulator)

# Define different shot counts (number of repetitions of the experiment)
shot_values = [10, 100, 1000, 5000]

# Prepare lists to store the measurement results and corresponding labels
data_for_plot = []
legends = []

# Execute the same circuit with different numbers of shots
for shots in shot_values:
    job = simulator.run(compiled_circuit, shots=shots)  # Run the circuit
    result = job.result()                               # Get the simulation result
    counts = result.get_counts()                        # Retrieve measurement counts
    data_for_plot.append(dict(counts))                  # Store the result as a dictionary
    legends.append(f"{shots} shots")                    # Create a label for the legend

# Plot all histograms together for comparison
fig = plot_histogram(
    data_for_plot,
    legend=legends,
    title="Effect of Shot Count on Measurement Statistics",
    figsize=(9, 5)
)

# Display the plot in the notebook
plt.show()
```

#### Theoretical Answer:

In quantum computing, a **shot** represents one complete execution of a quantum circuit — that is, the process of preparing the qubit, evolving it through quantum gates, and measuring the result once.

Because measurement outcomes are **probabilistic**, the true distribution of results can only be estimated by running the circuit multiple times. The larger the number of shots, the closer the experimental frequencies will approach the **theoretical probabilities**.

For a qubit in an equal superposition, like:

$$
|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle),
$$

the expected probabilities are:

$$
P(0) = 0.5, \quad P(1) = 0.5
$$

However, if we only perform a small number of measurements (for example, 10 shots), random fluctuations may cause results to deviate significantly from the expected 50–50 distribution. As we increase the number of shots to 1000 or 5000, the histogram gradually stabilizes, and the relative frequencies converge toward the true theoretical values.

This exercise demonstrates the importance of **statistical sampling** in quantum experiments, and how the apparent randomness of individual measurements reveals consistent probabilities when viewed collectively.

    <h3 style="margin-top:0;">Exercise 4: <strong>Modify</strong> the visualization code to display both the Bloch vector and the measurement histogram for the same circuit, and <strong>interpret</strong> their relationship.</h3>
</div>  

#### Code Answer:

```python
# --- Exercise 4: Visualizing the Bloch Vector and the Measurement Histogram ---

# Ensures plots are shown in Jupyter
%matplotlib inline

# Create a circuit with one qubit and one classical bit
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate to create a balanced superposition
qc.h(0)

# Measure the qubit
qc.measure(0, 0)

# Draw the circuit
qc.draw('mpl')

# Initialize the Aer simulator
simulator = AerSimulator()

# Transpile and execute the circuit
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()

# Get measurement counts
counts = result.get_counts()
print("Measurement results:", counts)

# --- Visualization ---

# Bloch vector for |+⟩ state (after Hadamard on |0⟩)
bloch_vector = [1, 0, 0]

# Create a figure with 2 subplots:
# - the first one will be 3D for the Bloch sphere
# - the second one will be 2D for the measurement histogram
fig = plt.figure(figsize=(10, 4))

# 1st subplot: 3D axis for the Bloch sphere
ax_bloch = fig.add_subplot(1, 2, 1, projection='3d')

# 2nd subplot: normal 2D axis for the histogram
ax_hist = fig.add_subplot(1, 2, 2)

# Plot Bloch vector on the 3D axis
plot_bloch_vector(bloch_vector, title="Qubit State on Bloch Sphere", ax=ax_bloch)

# Plot histogram of measurements on the 2D axis
plot_histogram(dict(counts), ax=ax_hist, title="Measurement Histogram")

# Slight manual adjustment to avoid overlap (instead of tight_layout, which fails with 3D)
plt.subplots_adjust(wspace=0.4)

# Show the final figure
plt.show()
```

#### Theoretical Answer:

The Bloch sphere and the measurement histogram are two complementary ways of understanding a qubit’s behavior.

The **Bloch sphere** provides a geometric representation of the qubit’s quantum state before measurement. The position of the state vector (also called the **Bloch vector**) determines the probabilities of obtaining each measurement result. For example:

- If the vector points to the **north pole (0,0,1)**, the qubit is in **|0⟩** and the measurement always yields 0.
- If it points to the **south pole (0,0,-1)**, the qubit is in **|1⟩** and the measurement always yields 1.
- If it lies along the **x-axis**, the qubit is in a superposition (|+⟩ or |−⟩) with 50% probability of each outcome.

The **histogram** shows the results after many measurements. Each bar represents the frequency of obtaining 0 or 1. The relative heights of these bars correspond to the probabilities encoded by the Bloch vector’s orientation.

By visualizing both representations together, we can directly connect the **geometric description (state vector)** and the **statistical outcomes (measurement probabilities)**. This dual perspective is essential for understanding how quantum states evolve and how measurements extract information from them.

#### Theoretical Answer:

Until now, all circuits have been executed on simulators, which perfectly mimic ideal quantum behavior. However, in real quantum computers, qubits are physical systems that suffer from **noise**, **decoherence**, and **readout errors**. Running the same circuit on a real backend allows us to observe these imperfections in practice.

In this exercise, we will execute a simple circuit consisting of a single qubit placed in superposition using a **Hadamard gate**, followed by a measurement. The expected theoretical probabilities are:

$$
P(0) = 0.5, \quad P(1) = 0.5
$$

When we run this on a simulator, the histogram will be perfectly balanced. On a real device, however, small deviations will appear due to the physical limitations of the hardware. These deviations are normal and illustrate an essential aspect of **quantum hardware realism**: measurement results are probabilistic not only due to quantum mechanics, but also because of hardware noise.

It is worth noting that modern IBM quantum processors—especially the more recent Heron-generation systems—are **highly calibrated and significantly less noisy** than earlier devices. For very simple circuits like this one (a single-qubit superposition with no entangling gates), the noise level is so low that the results may appear almost indistinguishable from those of the simulator. This is expected: one-qubit gates typically have extremely high fidelities, and readout errors are often only a few percent. The effects of noise become much more evident as we increase the depth of a circuit or introduce multi-qubit operations.

By comparing simulator and real-backend results, you can appreciate the difference between an **idealized quantum system** and a **real-world quantum processor**. This understanding is critical before progressing to more advanced topics such as **error mitigation** and **quantum circuit optimization**.
